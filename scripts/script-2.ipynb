## Блок 2

Из БВ FRED загрузите ряд `MORTGAGE30US` ([ссылка](https://fred.stlouisfed.org/series/MORTGAGE30US)) c **2000-01-01** по **2025-12-31**. Для этого ряда подгоните модель ARIMA порядка (2, 1, 2) **без константы/тренда**. Проведите тест на ARCH-эффекты, число лагов возьмите равным 5. В ответе укажите тестовую статистику и сделайте вывод. Ответ округлите до 3-х десятичных знаков. Уровень значимости 5.0%

Тестовая статистика = Ответ {$а} Вопрос 2

Вывод: Ответ {$а} Вопрос 2 есть гетероскедастичностьнет гетероскедастичности

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import ARIMA
from statsmodels.stats.diagnostic import het_arch

# Загрузка ряда MORTGAGE30US из FRED
y = web.DataReader(
    name="MORTGAGE30US",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# ARIMA(2, 1, 2) без константы/тренда
mod = ARIMA(
    y,
    order=(2, 1, 2),
    trend="n",
    missing="drop"
)

res = mod.fit()

# ARCH LM-тест, число лагов = 5
lm_stat, lm_pvalue, f_stat, f_pvalue = het_arch(
    res.resid,
    nlags=5
)

print("LM-статистика:", round(lm_stat, 3))
print("p-value:", round(lm_pvalue, 3))

if lm_pvalue < 0.05:
    print("Вывод: есть гетероскедастичность")
else:
    print("Вывод: нет гетероскедастичности")
```

## Блок 3

Из БВ FRED загрузите ряд `MORTGAGE30US` ([ссылка](https://fred.stlouisfed.org/series/MORTGAGE30US)) c **2000-01-01** по **2025-12-31**. Для **первой разности** этого ряда вычислите коэффициенты _r(3)_ и _rpart(3)_. Ответ округлите до 3-х десятичных знаков.

_r(3)_=Ответ {$а} Вопрос 3

_rpart(3)_=Ответ {$а} Вопрос 3

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import acf, pacf

# Загрузка ряда MORTGAGE30US из FRED
y = web.DataReader(
    name="MORTGAGE30US",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# Берём сам ряд как Series
y = y["MORTGAGE30US"]

# Первая разность
dy = y.diff().dropna()

# ACF и PACF до лага 3 включительно
acf_values = acf(dy, nlags=3)
pacf_values = pacf(dy, nlags=3)

# r(3) и rpart(3)
r3 = acf_values[3]
rpart3 = pacf_values[3]

print("r(3) =", round(r3, 3))
print("rpart(3) =", round(rpart3, 3))
```

## Блок 5

Из БВ FRED загрузите ряд `WBAA` ([ссылка](https://fred.stlouisfed.org/series/WBAA)) c **2000-01-01** по **2025-12-31**. Для этого ряда проведите KPSS (выбрав подходящий вариант с константой/трендом) и с **выбором числа лагов по умолчанию**. В ответе укажите тестовую статистику и сделайте вывод. Уровень значимости 1%. Ответ округлите до 3-х десятичных знаков.

Тестовая статистика = Ответ {$а} Вопрос 5

Критическое значение = Ответ {$а} Вопрос 5

Вывод: Ответ {$а} Вопрос 5 есть единичный кореньнет единичного корня

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import kpss

# Загрузка ряда WBAA из FRED
y = web.DataReader(
    name="WBAA",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# Берём сам ряд как Series
y = y["WBAA"]

# KPSS-тест:
# regression="c" - тест с константой, без линейного тренда
# nlags="auto" - выбор числа лагов по умолчанию
kpss_result = kpss(
    y,
    regression="c",
    nlags="auto"
)

kpss_stat = kpss_result[0]
pvalue = kpss_result[1]
used_lags = kpss_result[2]
crit_values = kpss_result[3]

print("Использованное число лагов:", used_lags)
print("Тестовая статистика:", round(kpss_stat, 3))
print("Критическое значение 1%:", round(crit_values["1%"], 3))
print("p-value:", round(pvalue, 3))

if kpss_stat > crit_values["1%"]:
    print("Вывод: есть единичный корень")
else:
    print("Вывод: нет единичного корня")
```

## Блок 6

Из БВ FRED загрузите ряд `MORTGAGE30US` ([ссылка](https://fred.stlouisfed.org/series/MORTGAGE30US)) c **2000-01-01** по **2025-12-31**. Для этого ряда подгоните модель ARIMA порядка (0, 1, 2) **без константы/тренда** и укажите её коэффициенты. Ответ округлите до 3-х десятичных знаков.

θ1=Ответ {$а} Вопрос 6

θ2=Ответ {$а} Вопрос 6

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import ARIMA

# Загрузка ряда MORTGAGE30US из FRED
y = web.DataReader(
    name="MORTGAGE30US",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# ARIMA(0, 1, 2) без константы/тренда
mod = ARIMA(
    y,
    order=(0, 1, 2),
    trend="n",
    missing="drop"
)

res = mod.fit()

# Все коэффициенты модели
print(res.params)

# MA-коэффициенты
theta1 = res.params["ma.L1"]
theta2 = res.params["ma.L2"]

print("θ1 =", round(theta1, 3))
print("θ2 =", round(theta2, 3))
```

## Блок 7

Из БВ FRED загрузите ряд `MORTGAGE30US` ([ссылка](https://fred.stlouisfed.org/series/MORTGAGE30US)) c **2010-01-01** по **2025-12-31**. Для этого ряда подгоните модель ARIMA **оптимального** порядка, используя тест единичного корня `adf` и информационный критерий `aicc`. Параметры `seasonal=False`, `max_p=3`, `max_q=3`. В ответе укажите порядок "оптимальной" модели.

p=Ответ {$а} Вопрос 7

d=Ответ {$а} Вопрос 7

q=Ответ {$а} Вопрос 7

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
import pmdarima as pm

# Загрузка ряда MORTGAGE30US из FRED
y = web.DataReader(
    name="MORTGAGE30US",
    data_source="fred",
    start="2010-01-01",
    end="2025-12-31"
).dropna()

# Берём сам ряд как Series
y = y["MORTGAGE30US"]

# Подбор оптимальной ARIMA
arima = pm.auto_arima(
    y,
    test="adf",
    information_criterion="aicc",
    seasonal=False,
    max_p=3,
    max_q=3,
    suppress_warnings=True,
    error_action="ignore",
    trace=True
)

print("Оптимальный порядок:", arima.order)

p, d, q = arima.order

print("p =", p)
print("d =", d)
print("q =", q)
```
