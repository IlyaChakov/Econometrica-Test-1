## Блок 1

Из БВ FRED загрузите ряд `WAAA` ([ссылка](https://fred.stlouisfed.org/series/WAAA)) c **2000-01-01** по **2025-12-31**. Для **первой разности** этого ряда проведите ADF (выбрав подходящий вариант с константой/трендом) и с **выбором числа лагов по умолчанию**. В ответе укажите тестовую статистику и сделайте вывод. Уровень значимости 5%. Ответ округлите до 3-х десятичных знаков.

Тестовая статистика = Ответ {$а} Вопрос 1

Критическое значение = Ответ {$а} Вопрос 1

Вывод: Ответ {$а} Вопрос 1 есть единичный кореньнет единичного корня

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import adfuller

# Загрузка ряда WAAA из FRED
y = web.DataReader(
    name="WAAA",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# Берём сам ряд как Series
y = y["WAAA"]

# Первая разность
dy = y.diff().dropna()

# ADF-тест:
# regression="n" - без константы и без тренда
# autolag="AIC" - выбор числа лагов по умолчанию
adf_result = adfuller(
    dy,
    regression="n",
    autolag="AIC"
)

adf_stat = adf_result[0]
pvalue = adf_result[1]
used_lag = adf_result[2]
crit_values = adf_result[4]

print("Использованное число лагов:", used_lag)
print("Тестовая статистика:", round(adf_stat, 3))
print("Критическое значение 5%:", round(crit_values["5%"], 3))
print("p-value:", round(pvalue, 3))

if adf_stat < crit_values["5%"]:
    print("Вывод: нет единичного корня")
else:
    print("Вывод: есть единичный корень")
```

## Блок 2

Из БВ FRED загрузите ряд `WAAA` ([ссылка](https://fred.stlouisfed.org/series/WAAA)) c **2010-01-01** по **2025-12-31**. Для этого ряда подгоните модель ARIMA **оптимального** порядка, используя тест единичного корня `adf` и информационный критерий `bic`. Параметры `seasonal=False`, `max_p=3`, `max_q=3`. В ответе укажите порядок "оптимальной" модели.

p=Ответ {$а} Вопрос 2

d=Ответ {$а} Вопрос 2

q=Ответ {$а} Вопрос 2

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
import pmdarima as pm

# Загрузка ряда WAAA из FRED
y = web.DataReader(
    name="WAAA",
    data_source="fred",
    start="2010-01-01",
    end="2025-12-31"
).dropna()

# Берём сам ряд как Series
y = y["WAAA"]

# Подбор оптимальной ARIMA
arima = pm.auto_arima(
    y,
    test="adf",
    information_criterion="bic",
    seasonal=False,
    max_p=3,
    max_q=3,
    suppress_warnings=True,
    error_action="ignore",
    trace=True
)

print("Оптимальный порядок:", arima.order)

p, d, q = arima.order

print("p =", p)
print("d =", d)
print("q =", q)
```

## Блок 3

Из БВ FRED загрузите ряд `WAAA` ([ссылка](https://fred.stlouisfed.org/series/WAAA)) c **2000-01-01** по **2025-12-31**. Для этого ряда подгоните модель ARIMA порядка (3, 2, 0) **без константы/тренда** и укажите её коэффициенты. Ответ округлите до 3-х десятичных знаков.

φ1=Ответ {$а} Вопрос 3

φ2=Ответ {$а} Вопрос 3

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import ARIMA

# Загрузка ряда WAAA из FRED
y = web.DataReader(
    name="WAAA",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# ARIMA(3, 2, 0) без константы/тренда
mod = ARIMA(
    y,
    order=(3, 2, 0),
    trend="n",
    missing="drop"
)

res = mod.fit()

# Все коэффициенты модели
print(res.params)

# AR-коэффициенты
phi1 = res.params["ar.L1"]
phi2 = res.params["ar.L2"]

print("φ1 =", round(phi1, 3))
print("φ2 =", round(phi2, 3))
```

## Блок 5

Из БВ FRED загрузите ряд `WBAA` ([ссылка](https://fred.stlouisfed.org/series/WBAA)) c **2000-01-01** по **2025-12-31**. Для **первой разности** этого ряда вычислите коэффициенты _r(1)_ и _rpart(1)_. Ответ округлите до 3-х десятичных знаков.

_r(1)_=Ответ {$а} Вопрос 5

_rpart(1)_=Ответ {$а} Вопрос 5

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import acf, pacf

# Загрузка ряда WBAA из FRED
y = web.DataReader(
    name="WBAA",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# Берём сам ряд как Series
y = y["WBAA"]

# Первая разность
dy = y.diff().dropna()

# ACF и PACF до лага 1 включительно
acf_values = acf(dy, nlags=1)
pacf_values = pacf(dy, nlags=1)

# r(1) и rpart(1)
r1 = acf_values[1]
rpart1 = pacf_values[1]

print("r(1) =", round(r1, 3))
print("rpart(1) =", round(rpart1, 3))
```

## Блок 6

Из БВ FRED загрузите ряд `WAAA` ([ссылка](https://fred.stlouisfed.org/series/WAAA)) c **2000-01-01** по **2025-12-31**. Для этого ряда подгоните модель ARIMA порядка (3, 2, 0) **без константы/тренда**. Проведите тест на серийную корреляцию, число лагов возьмите равным 6. В ответе укажите тестовую статистику и сделайте вывод. Ответ округлите до 3-х десятичных знаков. Уровень значимости 1.0%

Тестовая статистика = Ответ {$а} Вопрос 6

Вывод: Ответ {$а} Вопрос 6 есть серинйая корреляциянет серийной корреляции

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import ARIMA
from statsmodels.stats.diagnostic import acorr_ljungbox

# Загрузка ряда WAAA из FRED
y = web.DataReader(
    name="WAAA",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# ARIMA(3, 2, 0) без константы/тренда
mod = ARIMA(
    y,
    order=(3, 2, 0),
    trend="n",
    missing="drop"
)

res = mod.fit()

# Ljung-Box тест, число лагов = 6
lb_test = acorr_ljungbox(
    res.resid,
    lags=[6],
    return_df=True
)

print(lb_test)

stat = lb_test.loc[6, "lb_stat"]
pvalue = lb_test.loc[6, "lb_pvalue"]

print("Тестовая статистика:", round(stat, 3))
print("p-value:", round(pvalue, 3))

if pvalue < 0.01:
    print("Вывод: есть серийная корреляция")
else:
    print("Вывод: нет серийной корреляции")
```

## Блок 7

Из БВ FRED загрузите ряд `WBAA` ([ссылка](https://fred.stlouisfed.org/series/WBAA)) c **2000-01-01** по **2025-12-31**. Для этого ряда подгоните модель ARIMA порядка (1, 2, 1) **без константы/тренда**. По подогнанной модели постройте прогноз на 2 периода вперёд. Ответ округлите до 3-х десятичных знаков.

Прогноз на 1 период = Ответ {$а} Вопрос 7

Прогноз на 2 период = Ответ {$а} Вопрос 7

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import ARIMA

# Загрузка ряда WBAA из FRED
y = web.DataReader(
    name="WBAA",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# ARIMA(1, 2, 1) без константы/тренда
mod = ARIMA(
    y,
    order=(1, 2, 1),
    trend="n",
    missing="drop"
)

res = mod.fit()

# Прогноз на 2 периода вперёд
forecast = res.forecast(steps=2)

print(forecast.round(3))
```
