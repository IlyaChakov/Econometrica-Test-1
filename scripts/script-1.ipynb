## Блок 1

Из БВ FRED загрузите ряд `WAAA` ([ссылка](https://fred.stlouisfed.org/series/WAAA)) c **2000-01-01** по **2025-12-31**. Для этого ряда подгоните модель ARIMA порядка (1, 0, 3) **без константы/тренда**. По подогнанной модели постройте прогноз на 2 периода вперёд. Ответ округлите до 3-х десятичных знаков.

Прогноз на 1 период = Ответ {$а} Вопрос 1

Прогноз на 2 период = Ответ {$а} Вопрос 1

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import ARIMA

# Загрузка ряда WAAA из FRED
y = web.DataReader(
    name="WAAA",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# ARIMA(1, 0, 3) без константы/тренда
mod = ARIMA(
    y,
    order=(1, 0, 3),
    trend="n",
    missing="drop"
)

res = mod.fit()

# Прогноз на 2 периода вперёд
forecast = res.forecast(steps=2)

# Округление до 3 знаков
forecast_rounded = forecast.round(3)

print(forecast_rounded)
```

## Блок 2

Из БВ FRED загрузите ряд `MORTGAGE30US` ([ссылка](https://fred.stlouisfed.org/series/MORTGAGE30US)) c **2000-01-01** по **2025-12-31**. Для этого ряда подгоните модель ARIMA порядка (2, 1, 2) **без константы/тренда**. Проведите тест на серийную корреляцию, число лагов возьмите равным 5. В ответе укажите тестовую статистику и сделайте вывод. Ответ округлите до 3-х десятичных знаков. Уровень значимости 5.0%

Тестовая статистика = Ответ {$а} Вопрос 2

Вывод: Ответ {$а} Вопрос 2 есть серинйая корреляциянет серийной корреляции

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import ARIMA
from statsmodels.stats.diagnostic import acorr_ljungbox

# Загрузка ряда из FRED
y = web.DataReader(
    name="MORTGAGE30US",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# ARIMA(2, 1, 2) без константы/тренда
mod = ARIMA(
    y,
    order=(2, 1, 2),
    trend="n",
    missing="drop"
)

res = mod.fit()

# Тест Ljung-Box на серийную корреляцию, lag = 5
lb_test = acorr_ljungbox(res.resid, lags=[5], return_df=True)

print(lb_test)

stat = lb_test.loc[5, "lb_stat"]
pvalue = lb_test.loc[5, "lb_pvalue"]

print("Тестовая статистика:", round(stat, 3))
print("p-value:", round(pvalue, 3))

if pvalue < 0.05:
    print("Вывод: есть серийная корреляция")
else:
    print("Вывод: нет серийной корреляции")
```

## Блок 3

Из БВ FRED загрузите ряд `WAAA` ([ссылка](https://fred.stlouisfed.org/series/WAAA)) c **2000-01-01** по **2025-12-31**. Для этого ряда проведите ADF (выбрав подходящий вариант с константой/трендом) и с **выбором числа лагов по умолчанию**. В ответе укажите тестовую статистику и сделайте вывод. Уровень значимости 5%. Ответ округлите до 3-х десятичных знаков.

Тестовая статистика = Ответ {$а} Вопрос 3

Критическое значение = Ответ {$а} Вопрос 3

Вывод: Ответ {$а} Вопрос 3 есть единичный кореньнет единичного корня

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import adfuller

# Загрузка ряда WAAA из FRED
y = web.DataReader(
    name="WAAA",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# Берём сам ряд как Series
y = y["WAAA"]

# ADF-тест:
# regression="c" означает тест с константой, без линейного тренда
# autolag="AIC" - выбор числа лагов по умолчанию в statsmodels
adf_result = adfuller(
    y,
    regression="c",
    autolag="AIC"
)

adf_stat = adf_result[0]
pvalue = adf_result[1]
used_lag = adf_result[2]
crit_values = adf_result[4]

print("Использованное число лагов:", used_lag)
print("Тестовая статистика:", round(adf_stat, 3))
print("Критическое значение 5%:", round(crit_values["5%"], 3))
print("p-value:", round(pvalue, 3))

if adf_stat < crit_values["5%"]:
    print("Вывод: нет единичного корня")
else:
    print("Вывод: есть единичный корень")
```

## Блок 4

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import ARIMA

# Загрузка ряда MORTGAGE30US из FRED
y = web.DataReader(
    name="MORTGAGE30US",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# ARIMA(2, 1, 2) без константы/тренда
mod = ARIMA(
    y,
    order=(2, 1, 2),
    trend="n",
    missing="drop"
)

res = mod.fit()

# Все коэффициенты модели
print(res.params)

# Коэффициенты AR-части
phi1 = res.params["ar.L1"]
phi2 = res.params["ar.L2"]

print("φ1 =", round(phi1, 3))
print("φ2 =", round(phi2, 3))
```

## Блок 6

Из БВ FRED загрузите ряд `WAAA` ([ссылка](https://fred.stlouisfed.org/series/WAAA)) c **2000-01-01** по **2025-12-31**. Для **первой разности** этого ряда вычислите коэффициенты _r(2)_ и _rpart(2)_. Ответ округлите до 3-х десятичных знаков.

_r(2)_=Ответ {$а} Вопрос 6

_rpart(2)_=Ответ {$а} Вопрос 6

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
from statsmodels.tsa.api import acf, pacf

# Загрузка ряда WAAA из FRED
y = web.DataReader(
    name="WAAA",
    data_source="fred",
    start="2000-01-01",
    end="2025-12-31"
).dropna()

# Берём сам ряд как Series
y = y["WAAA"]

# Первая разность
dy = y.diff().dropna()

# ACF и PACF до лага 2 включительно
acf_values = acf(dy, nlags=2)
pacf_values = pacf(dy, nlags=2)

# r(2) и rpart(2)
r2 = acf_values[2]
rpart2 = pacf_values[2]

print("r(2) =", round(r2, 3))
print("rpart(2) =", round(rpart2, 3))
```

## Блок 7

Из БВ FRED загрузите ряд `WBAA` ([ссылка](https://fred.stlouisfed.org/series/WBAA)) c **2010-01-01** по **2025-12-31**. Для этого ряда подгоните модель ARIMA **оптимального** порядка, используя тест единичного корня `kpss` и информационный критерий `aic`. Параметры `seasonal=False`, `max_p=3`, `max_q=3`. В ответе укажите порядок "оптимальной" модели.

p=Ответ {$а} Вопрос 7

d=Ответ {$а} Вопрос 7

q=Ответ {$а} Вопрос 7

```python
import warnings
warnings.filterwarnings("ignore")

import pandas_datareader.data as web
import pmdarima as pm

# Загрузка ряда WBAA из FRED
y = web.DataReader(
    name="WBAA",
    data_source="fred",
    start="2010-01-01",
    end="2025-12-31"
).dropna()

# Берём сам ряд как Series
y = y["WBAA"]

# Подбор оптимальной ARIMA
arima = pm.auto_arima(
    y,
    test="kpss",
    information_criterion="aic",
    seasonal=False,
    max_p=3,
    max_q=3,
    suppress_warnings=True,
    error_action="ignore",
    trace=True
)

print("Оптимальный порядок:", arima.order)

p, d, q = arima.order

print("p =", p)
print("d =", d)
print("q =", q)
```
