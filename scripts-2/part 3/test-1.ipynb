## Задание 1

Из БВ FRED загрузите ряд `NA000254Q` ([ссылка](https://fred.stlouisfed.org/series/NA000254Q)) c **2002-01-01** по **2025-12-31**. Для **логарифма** этого ряда подгоните модель SARIMA (2, 0, 0)(0, 1, 1, 4) **с трендом**. По подогнанной модели постройте прогноз на 2 периода вперёд. Ответ округлите до 3-х десятичных знаков. **Указание**: для подгоки модели используйте класс `ARIMA`.

Прогноз на 1 период = Ответ {$а} Вопрос 1

Прогноз на 2 период = Ответ {$а} Вопрос 1

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
from pandas_datareader.data import DataReader
from statsmodels.tsa.arima.model import ARIMA

x = DataReader(
    "NA000254Q",
    "fred",
    start="2002-01-01",
    end="2025-12-31",
)["NA000254Q"].dropna()

# Ряд квартальный
x.index = x.index.to_period("Q")

# Логарифм ряда
y = np.log(x)

model = ARIMA(
    y,
    order=(2, 0, 0),
    seasonal_order=(0, 1, 1, 4),
    trend="t",   # тренд; при d=1 это аналог drift
)

res = model.fit()

forecast = res.forecast(steps=2)

print(forecast)
print(round(forecast.iloc[0], 3))
print(round(forecast.iloc[1], 3))
```
