## Задание 2

Из БВ FRED загрузите ряд `WBAA` ([ссылка](https://fred.stlouisfed.org/series/WBAA)) c **2010-01-01** по **2025-12-31**. Для **первой разности** этого ряда подгоните модель AR(2)-GARCH(2, 2, 1) c λ=1 и укажите её коэффициенты. Ответ округлите до 3-х десятичных знаков.

α1=Ответ {$а} Вопрос 2

β1=Ответ {$а} Вопрос 2

```python
import warnings
warnings.filterwarnings("ignore")

from pandas_datareader.data import DataReader
from arch import arch_model

x = DataReader(
    "WBAA",
    "fred",
    start="2010-01-01",
    end="2025-12-31",
)["WBAA"].dropna()

y = x.diff().dropna()

model = arch_model(
    y,
    mean="ARX",
    lags=2,
    vol="GARCH",
    p=2,
    o=2,
    q=1,
    power=1.0,
    dist="normal",
    rescale=False,
)

res = model.fit(disp="off")

print(res.params)

alpha1 = res.params["alpha[1]"]
beta1 = res.params["beta[1]"]

print(round(alpha1, 3))
print(round(beta1, 3))
```

## Задание 3

Из БВ FRED загрузите ряд `WAAA` ([ссылка](https://fred.stlouisfed.org/series/WAAA)) c **2010-01-01** по **2025-12-31**. Для **первой разности** этого ряда подгоните модель AR(3)-EGARCH(1, 1, 2). По подогнанной модели постройте прогноз на 1 период вперёд. В ответе укажите значение прогноза, умноженного на **1000**. Ответ округлите до 3-х десятичных знаков.

Прогноз на 1 период = Ответ {$а} Вопрос 3

```python
import warnings
warnings.filterwarnings("ignore")

from pandas_datareader.data import DataReader
from arch import arch_model

x = DataReader(
    "WAAA",
    "fred",
    start="2010-01-01",
    end="2025-12-31",
)["WAAA"].dropna()

y = x.diff().dropna()

model = arch_model(
    y,
    mean="ARX",
    lags=3,
    vol="EGARCH",
    p=1,
    o=1,
    q=2,
    dist="normal",
    rescale=False,
)

res = model.fit(disp="off")

forecast = res.forecast(horizon=1, reindex=False)
f1 = forecast.mean.iloc[-1, 0]

answer = f1 * 1000
print(round(answer, 3))
```

## Задание 4

Из БВ FRED загрузите ряд `MORTGAGE15US` ([ссылка](https://fred.stlouisfed.org/series/MORTGAGE15US)) c **2010-01-01** по **2025-12-31**. Для **первой разности** этого ряда подгоните модель AR(1)-GARCH(1, 1, 1) c λ=2. По подогнанной модели постройте прогноз на 2 период вперёд. В ответе укажите значение прогноза, умноженного на **1000**. Ответ округлите до 3-х десятичных знаков.

Прогноз на 1 период = Ответ {$а} Вопрос 4

Прогноз на 2 период = Ответ {$а} Вопрос 4

```python
import warnings
warnings.filterwarnings("ignore")

from pandas_datareader.data import DataReader
from arch import arch_model

x = DataReader(
    "MORTGAGE15US",
    "fred",
    start="2010-01-01",
    end="2025-12-31",
)["MORTGAGE15US"].dropna()

y = x.diff().dropna()

model = arch_model(
    y,
    mean="ARX",
    lags=1,
    vol="GARCH",
    p=1,
    o=1,
    q=1,
    power=2.0,      # lambda = 2
    dist="normal",
    rescale=False,
)

res = model.fit(disp="off")

forecast = res.forecast(horizon=2, reindex=False)
means = forecast.mean.iloc[-1]

h1 = means.iloc[0] * 1000
h2 = means.iloc[1] * 1000

print(round(h1, 3))
print(round(h2, 3))
```
