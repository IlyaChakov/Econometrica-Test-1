## Задание 1

Из БВ FRED загрузите ряд `WBAA` ([ссылка](https://fred.stlouisfed.org/series/WBAA)) c **2010-01-01** по **2025-12-31**. Для **первой разности** этого ряда подгоните модель AR(2)-EGARCH(2, 2, 1). По подогнанной модели постройте прогноз на 1 период вперёд. В ответе укажите значение прогноза, умноженного на **1000**. Ответ округлите до 3-х десятичных знаков.

Прогноз на 1 период = 

```python
import warnings
warnings.filterwarnings("ignore")

from pandas_datareader.data import DataReader
from arch import arch_model

# 1. Загружаем ряд WBAA
x = DataReader(
    "WBAA",
    "fred",
    start="2010-01-01",
    end="2025-12-31",
)["WBAA"].dropna()

# 2. Первая разность
y = x.diff().dropna()

# 3. AR(2)-EGARCH(2, 2, 1)
model = arch_model(
    y,
    mean="ARX",
    lags=2,
    vol="EGARCH",
    p=2,
    o=2,
    q=1,
    dist="normal",
    rescale=False,
)

res = model.fit(disp="off")

# 4. Прогноз на 1 период вперед
forecast = res.forecast(horizon=1, reindex=False)
f1 = forecast.mean.iloc[-1, 0]

answer = f1 * 1000
print(round(answer, 3))
```

## Задание 2 -

Из БВ FRED загрузите ряд `MORTGAGE30US` ([ссылка](https://fred.stlouisfed.org/series/MORTGAGE30US)) c **2010-01-01** по **2025-12-31**. Для **первой разности** этого ряда подгоните модель AR(2)-GARCH(1, 0, 1) c λ=2. По подогнанной модели постройте прогноз на 2 период вперёд. В ответе укажите значение прогноза, умноженного на **1000**. Ответ округлите до 3-х десятичных знаков.

Прогноз на 1 период = 

Прогноз на 2 период = 

```python
import warnings
warnings.filterwarnings("ignore")

from pandas_datareader.data import DataReader
from arch import arch_model

x = DataReader(
    "MORTGAGE30US",
    "fred",
    start="2010-01-01",
    end="2025-12-31",
)["MORTGAGE30US"].dropna()

y = x.diff().dropna()

model = arch_model(
    y,
    mean="ARX",
    lags=3, # Нужно было поставить 2 лага, а не 3
    vol="GARCH",
    p=1,
    o=0,
    q=1,
    power=2.0,      # lambda = 2
    dist="normal",
    rescale=False,
)

res = model.fit(disp="off")

forecast = res.forecast(horizon=2, reindex=False)
means = forecast.mean.iloc[-1]

h1 = means.iloc[0] * 1000
h2 = means.iloc[1] * 1000

print(round(h1, 3))
print(round(h2, 3))
```

## Задание 4

Из БВ FRED загрузите ряд `MORTGAGE15US` ([ссылка](https://fred.stlouisfed.org/series/MORTGAGE15US)) c **2010-01-01** по **2025-12-31**. Для **первой разности** этого ряда подгоните модель AR(1)-EGARCH(1, 1, 1) и укажите её коэффициенты. Ответ округлите до 3-х десятичных знаков.

α1 =

β1 =

```python
import warnings
warnings.filterwarnings("ignore")

from pandas_datareader.data import DataReader
from arch import arch_model

x = DataReader(
    "MORTGAGE15US",
    "fred",
    start="2010-01-01",
    end="2025-12-31",
)["MORTGAGE15US"].dropna()

y = x.diff().dropna()

model = arch_model(
    y,
    mean="ARX",
    lags=1,
    vol="EGARCH",
    p=1,
    o=1,
    q=1,
    dist="normal",
    rescale=False,
)

res = model.fit(disp="off")

print(res.params)

alpha1 = res.params["alpha[1]"]
beta1 = res.params["beta[1]"]

print(round(alpha1, 3))
print(round(beta1, 3))
```
